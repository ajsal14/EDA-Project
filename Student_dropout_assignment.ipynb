```python
import pandas as pd
df=pd.read_csv("student_dropout[1].csv")
df.head()
```

```python
print(df.shape)
print(df.dtypes)
df.describe()
```

```python
missing_values=df.isnull().sum()
print(missing_values)
```

```python
df['Age']=df['Age'].fillna(df['Age'].median()) # Changed 'age' to 'Age'
```

```python
for col in ['Address','Gender','Final_Grade']:
  print(df[col].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Address',data=df)
plt.title('Distribution of Address Status')
plt.show()
```

```python
sns.histplot(x='Age',data=df)
plt.title('Age Distribution')
plt.show()
```

```python
sns.boxplot(x='Final_Grade',data=df)
plt.title('Final Grade boxplot')
plt.show()
```

```python
sns.countplot(x='Gender',data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
df['Gender'] = df['Gender'].map({'M': 0, 'F': 1})
Connection_Matrix = df.corr(numeric_only=True) # Add numeric_only=True
plt.figure(figsize=(10, 6))
sns.heatmap(Connection_Matrix, annot=True, cmap='coolwarm') # Change Correlation_Matrix to Connection_Matrix
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(df[['Age','Address','Number_of_Absences','Final_Grade']],hue='Address')

plt.show()
```

```python
sns.boxplot(x='Address',y='Final_Grade',data=df)
plt.title('Final Grade vs Address')
plt.show()
```
